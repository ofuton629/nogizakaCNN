```python
import torch
import numpy as np
import torch.nn as nn
import matplotlib.pyplot as plt
import japanize_matplotlib
from torchviz import make_dot
from torchinfo import summary
from IPython.display import display
from tqdm.notebook import tqdm
import torch.optim as optim
```

```python
#学習率
a=0.1

#関数決定　y=x**2+p*x+q
p=-6
q=8

#初期点
x=25

#初期値
m=0
v=0
t=0    

b1_list=[]
b2_list=[]
b3_list=[]
b4_list=[]
b5_list=[]
b6_list=[]
b7_list=[]
b8_list=[]
b9_list=[]
b10_list=[]


for b1 in np.arange(0,1,0.01):
    
    for b2 in np.arange(0,1,0.01):
        
        x=25
        m=0
        v=0
        t=0
        
        while (abs(x+(p/2))>10**(-6))and(t<200000):
            
            t=t+1
            g=2*x+p
            m=b1*m+(1-b1)*g
            v=b2*v+(1-b2)*g**2
            mh=m/(1-b1**t)
            vh=v/(1-b2**t)
            x=x-a*(mh/np.sqrt(vh))
    
    
        if t<1000 :
            
            b1_list.append(b1)
            b2_list.append(b2)
        
        else:
            if t<5000:
                b3_list.append(b1)
                b4_list.append(b2)
                    
            else:
                if t<20000:
                    b5_list.append(b1)
                    b6_list.append(b2)
                            
                else:
                    if t<100000:
                        b7_list.append(b1)
                        b8_list.append(b2)
                                    
                    else:
                        if t<200000:
                            b9_list.append(b1)
                            b10_list.append(b2)                
```

```python
print(b1_list)
print(b2_list)
```

```python
fig = plt.figure(figsize=(30, 15))

i=1

ax = fig.add_subplot(2,3, i)


b1=np.linspace(0,1,100)

b2=0.3*b1+0.7

ax.plot(b1,b2,color="k")

ax.scatter(b1_list, b2_list,s=20,c='darkviolet',label="t<1000")
ax.scatter(b3_list, b4_list,s=20,c='b',label="1000≦t<5000")
ax.scatter(b5_list, b6_list,s=20,c='lime',label="5000≦t<20000")
ax.scatter(b7_list, b8_list,s=20,c='orange',label="20000≦t<100000")
ax.scatter(b9_list, b10_list,s=20,c='red',label="100000≦t≦200000")


ax.set_title('Adamのパラメータ別収束までの反復回数')
ax.set_xlabel('b1', fontsize=16)
ax.set_ylabel('b2', fontsize=16)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.tight_layout() 
plt.legend()
plt.show()
```
