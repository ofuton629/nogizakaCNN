```python
#SGD（momentum）を用いたx^2+y^2の最小値
```

```python
import torch
from torch import optim
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_list=[]
y_list=[]
f_list=[]
```

```python
def func(x,y):
    return x**2+y**2
```

```python
x = torch.tensor(-75.0, requires_grad=True)
y = torch.tensor(-10.0, requires_grad=True)
params = [x, y]
    
lr = 1

optimizer=optim.SGD(params, lr=0.1,momentum=0.9)
    

for i in range(1000):
    optimizer.zero_grad()
    outputs=func(x, y)
    outputs.backward()
    optimizer.step()
    x_list.append(x.item())
    y_list.append(y.item())
    f_list.append(outputs.item())
```

```python
fig = plt.figure(figsize=(15, 10))

i=1

ax = fig.add_subplot(2,3, i)
ax.scatter(x_list, y_list)
ax.plot(-75, -10, '+')
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel('y', fontsize=16)
ax.set_xlim(-80, 80)
ax.set_ylim(-15, 15)

x1 = np.arange(-80, 80, 0.01) 
y1 = np.arange(-15, 15, 0.01) 

X, Y = np.meshgrid(x1, y1)
Z =func(X,Y)

cont = plt.contour(X, Y, Z)


    
plt.tight_layout()    
fig.show()
```
